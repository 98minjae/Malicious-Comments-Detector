```python
##Pre-Processing Training Dataset

train['Comment'] = train['Comment'].apply(lambda x: " ".join(x.lower() for x in x.split()))
##Lowercasing
train['Comment'] = train['Comment'].str.replace('[^\w\s]','')
##Punctuation Removal
from textblob import Word
train['Comment'] = train['Comment'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
##Lemmatization
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['Comment'] = train['Comment'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
##Stopwords Removal
```

```python
##Pre-Pressing Predicting Dataset

predict['Comment'] = predict['Comment'].apply(lambda x: " ".join(x.lower() for x in x.split()))
##Lowercasing
predict['Comment'] = predict['Comment'].str.replace('[^\w\s]','')
##Punctuation Removal
from textblob import Word
predict['Comment'] = predict['Comment'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
##Lemmatization
from nltk.corpus import stopwords
stop = stopwords.words('english')
predict['Comment'] = predict['Comment'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
##Stopwords Removal
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
models = [
    SVC(),
    LogisticRegression(random_state=0),
    DecisionTreeClassifier()
]
X = train['Comment']
y = train['Type']
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0, random_state=0)

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)

CV = 10
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
  model_name = model.__class__.__name__
  accuracies = cross_val_score(model, X=X_train, y=y_train, cv=CV)
  for fold_idx, accuracy in enumerate(accuracies):
    entries.append((model_name, fold_idx, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])

cv_df.groupby('model_name').accuracy.mean()


```

```python
models[0].fit(X_train, y_train)
models[1].fit(X_train, y_train)
models[2].fit(X_train, y_train)


Xnew = predict['Comment']
vectorizer = TfidfVectorizer()
Xnew = vectorizer.fit_transform(Xnew)


ynew = models[0].predict(Xnew)
ynew2 = models[1].predict(Xnew)
ynew3 = models[2].predict(Xnew)

for i in range(len(Xnew)):
    print("SVM:", Xnew[i], ynew[i])
    print("Logistic Regression:", Xnew[i], ynew2[i])
    print("Decision Tree:", Xnew[i], ynew3[i])
```
